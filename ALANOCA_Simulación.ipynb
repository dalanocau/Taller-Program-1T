```python
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
import pandas as pd
```

```python
# Parámetros
n_clientes = 150      # Número de clientes seleccionados aleatoriamente
p_queja = 0.018       # Probabilidad de que un cliente tenga una queja
DNI = 152    # Semilla para la simulación
n_transacciones = 500 #Número de Transacciones
```

```python
# a) Simular 700 valores de la cantidad de quejas por problemas de facturación entre 150 clientes seleccionados
distrib = st.binom(n=n_clientes, p=p_queja)
simulacion_700 = distrib.rvs(size=700, random_state=DNI)

#Guardar simulacion_700 en data
data = pd.Series(simulacion_700).value_counts().sort_index()

#Usamos matplotlib y seaborn para la grafica
plt.figure(figsize=(10, 6))
sns.barplot(x=data.index, y=data.values, color='blue')

# Grafica
plt.title('Simulación de la Cantidad de Quejas por Problemas de Facturación (700 Simulaciones)', fontsize=14)
plt.xlabel('Cantidad de Quejas', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# b) Cálculo de la media y el desvío estándar de la simulación
media_simulacion = np.mean(simulacion_700)
desviacion_simulacion = np.std(simulacion_700)

# Cálculo de los valores teóricos
media_teorica = distrib.mean()
desviacion_teorica = distrib.std()

# Imprimir resultados
print(f'Media simulada: {media_simulacion:.2f}')
print(f'Desviación estándar simulada: {desviacion_simulacion:.2f}')
print(f'Media teórica: {media_teorica:.2f}')
print(f'Desviación estándar teórica: {desviacion_teorica:.2f}')
```

```python
# c) Simular 6000 valores de la cantidad de quejas por problemas de facturación entre 500 transacciones
distrib_500 = st.binom(n=n_transacciones, p=p_queja)
simulacion_6000 = distrib_500.rvs(size=6000, random_state=DNI)

# Mostrar los primeros 40 valores simulados
print(f'Primeros 40 valores simulados: {simulacion_6000[:40]}')
```

```python
# d) Verificar cuántas veces la cantidad de quejas es al menos 20
veces_mayor_o_igual_20 = np.sum(simulacion_6000 >= 20)
print(f'Cantidad de veces que las quejas son al menos 20: {veces_mayor_o_igual_20}')
```

```python
# e) Estimar la probabilidad de que entre 500 transacciones se generen al menos 20 quejas
probabilidad_simulada = veces_mayor_o_igual_20 / 6000
print(f'Probabilidad simulada de que haya al menos 20 quejas: {probabilidad_simulada:.4f}')

# Cálculo de la probabilidad teórica de que haya al menos 20 quejas cdf funcion de distribución acumulativa
probabilidad_teorica = 1 - distrib_500.cdf(19)
print(f'Probabilidad teórica de que haya al menos 20 quejas: {probabilidad_teorica:.4f}')
```
