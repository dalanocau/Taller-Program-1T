```python
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom  # Para cálculos teóricos y generación de muestras
```

```python
# Parámetros
n_clientes = 150      # Número de clientes seleccionados
p_queja = 0.018       # Probabilidad de que un cliente tenga una queja
DNI = 152    # Semilla para la simulación
```

```python
# a) Simular 700 valores de la cantidad de quejas por problemas de facturación entre 150 clientes seleccionados
distrib = binom(n=n_clientes, p=p_queja)  # Definir distribución binomial
simulacion_700 = distrib.rvs(size=700, random_state=DNI)

# Gráfico histograma de la simulación
plt.figure(figsize=(10, 6))
plt.hist(simulacion_700, bins=range(min(simulacion_700), max(simulacion_700) + 2),
         edgecolor='black', alpha=0.7, color='blue')
plt.title('Simulación de la Cantidad de Quejas por Problemas de Facturación (700 Simulaciones)', fontsize=14)
plt.xlabel('Cantidad de Quejas', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# b) Cálculo de la media y el desvío estándar de la simulación
media_simulacion = np.mean(simulacion_700)
desviacion_simulacion = np.std(simulacion_700)

# Cálculo de los valores teóricos
media_teorica = distrib.mean()
desviacion_teorica = distrib.std()

# Imprimir resultados
print(f'Media simulada: {media_simulacion:.2f}')
print(f'Desviación estándar simulada: {desviacion_simulacion:.2f}')
print(f'Media teórica: {media_teorica:.2f}')
print(f'Desviación estándar teórica: {desviacion_teorica:.2f}')
```
